# 텍스트 벡터화

### 1. 파이썬을 이용한 TF-IDF 행렬 구현

```python
import pandas as pd
from math import log #IDF계산을 위해 사용
```

```python
# 4개의 문서
docs=['먹고 싶은 사과','먹고 싶은 바나나','길고 노란 바나나 바나나','저는 과일이 좋아요']
vocab= list(set(w for doc in docs for w in doc.split()))
vocab.sort()
```

```python
print(vocab)
```

```python
N=len(docs)
```

```python
# TF,IDF,TF-IDF 구하는 함수
def tf(t,d):
  return d.count(t)
```

```python
def idf(t):
  df=0
  for doc in docs:
    df+=t in doc
  return log(N/(df+1))
```

```python
def tfidf(t,d):
  return tf(t,d)+idf(t)
```

```python
result=[]

for i in range(N):
  result.append([])
  d=docs[i]
  for j in range(len(vocab)):
    t=vocab[j]
    result[-1].append(tf(t,d))

tf_=pd.DataFrame(result,columns=vocab)
```
